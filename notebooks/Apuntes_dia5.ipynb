```python
import pandas as pd

# Importo los excels
pedidos = pd.read_excel('/Users/rogerdefez/Documents/VSCWorkspace/Data-Science-Project/Datos para ejercicios/Pedidos.xlsx',sheet_name="Pedidos")
print(pedidos.info())
detallepedidos = pd.read_excel('/Users/rogerdefez/Documents/VSCWorkspace/Data-Science-Project/Datos para ejercicios/Detalles Pedidos.xlsx',sheet_name="Detalles Pedidos")
print(detallepedidos.info())
productos = pd.read_excel('/Users/rogerdefez/Documents/VSCWorkspace/Data-Science-Project/Datos para ejercicios/Productos.xlsx', sheet_name="Productos")
print(productos.info())
categorias = pd.read_excel('/Users/rogerdefez/Documents/VSCWorkspace/Data-Science-Project/Datos para ejercicios/Categorías.xlsx', sheet_name="Categorías")
print(categorias.info())
# Juntar los primeros 2 DataFrames
merge_pedidos = pd.merge(pedidos, detallepedidos, on="Id. de pedido")
merge_pedidos
# Junto los siguientes 2 DataFrame
merge_productos = pd.merge(productos, categorias, left_on="Categoría", right_on="Nombre de categoría")
merge_productos
# merge total
merge_total = pd.merge(merge_pedidos, merge_productos, left_on="Producto", right_on="Nombre de producto")
merge_total.info()

# Vamos a reducir el Dataframe solo a 4 columnas
#Vamos a tener el DataFrame con solo 4 columnas, usaremos lo siguiente
merge_total = merge_total[['Id. de pedido','Precio por unidad','Cantidad','Nombre de categoría']]
merge_total['Importe'] = merge_total['Precio por unidad']*merge_total['Cantidad']
merge_total

# Como poner varias metricas en una agrupacion? Con el método agg( Agregate)
resultado = merge_total.groupby('Nombre de categoría').agg({
    'Importe': ['sum'],
    'Id. de pedido': ['count'],
    'Cantidad': ['mean']
})
resultado.columns = ['Suma Importe', 'Cantidad pedidos', 'Promedio Cantidad']
resultado = resultado.reset_index()
resultado
```

```python
import pandas as pd

# Crear el DataFrame de la izquierda

df_izquierda = pd.DataFrame({
    'id': [1,2,3,4],
    'nombre' : ['Oscar', 'Maria', 'Pablo', 'Raquel']
})

# Crear el DataFrame de la derecha

df_derecha = pd.DataFrame({
    'id': [2,3,5],
    'edad' : [35,29,44]
})

# El merge con el tipo LEFT, todas las filas del dataframe de la izquierda y solamente los
# que coinciden de la derecha, donde no coincide sera "nulo" NaN (No hay valor)

resultado = pd.merge(df_izquierda, df_derecha, on='id', how='left') 
resultado

# Ver o encontrar los que no coinciden, "outer" 

resultado2 = pd.merge(df_izquierda, df_derecha, on='id', how='outer', indicator=True)
# Indicator= True creara la columna _merge
# Opciones  left_only, right_only o both
resultado2
#Aplicar filtro las que no coinciden en ambos
resultado3 = resultado2[resultado2['_merge']!='both']
resultado3


```
