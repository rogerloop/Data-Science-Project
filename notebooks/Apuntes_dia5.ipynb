```python
#Como cargar un fichero de Excel, necesitaremos la libreria openpyxl
import pandas as pd
fichero = '/Users/rogerdefez/Documents/VSCWorkspace/Data-Science-Project/Datos para ejercicios/VentasGroucery_Meses.xlsx'
hojaAgosto = "Ventas Agosto 2014"
Agosto2014 = pd.read_excel(fichero,sheet_name=hojaAgosto)
Agosto2014
# LOS OTROS MESES SEPTIEMBRE OCTUBRE
hojaSeptiembre = "Ventas Septiembre 2014"
Septiembre2014 = pd.read_excel(fichero,sheet_name=hojaSeptiembre)

hojaOctubre = "Ventas Octubre 2014"
Octubre2014 = pd.read_excel(fichero,sheet_name=hojaOctubre)


#Ventas Agosto 2014
#"C:\Users\formacio\Downloads\VentasGroucery_Meses.xlsx"
# Anexar para tener un solo Dataframe
trimestre2014 = pd.concat([Agosto2014, Septiembre2014, Octubre2014], ignore_index=True)


# Ahora vamos a calcular el Amount = UnitPrice * Quantity

trimestre2014['Amount'] = trimestre2014['Unit Price'] * trimestre2014['Quantity']
trimestre2014

#

# la suma de Amount por pais
trimestre2014.groupby(['Ship Country']).Amount.sum()
# Promedio de precio  unidad por pais !!! Si el nombre tiene espacio 
# en blanco no se puede llamar con el punto hay que hacerlo con corchete
trimestre2014.groupby(['Ship City'])['Unit Price'].mean()

# El número de pedidos por categoria
trimestre2014.groupby(['Category'])['Order ID'].count()

#Vamos a agrupar por Pais y ciudad
trimestre2014.groupby(['Ship Country', 'Ship City'])['Amount'].sum()

# Suma por Categoria
Category_Amount = trimestre2014.groupby(['Category'])['Amount'].sum()
Category_Amount    # Este objeto no es un dataframe
Category_Amount = pd.DataFrame(Category_Amount, columns=['Amount'])
Category_Amount 

# Para generear el indice numerico al a vez que Category se convierte en columno
Category_Amount= Category_Amount.reset_index()
Category_Amount

# Si voy a usar estaos datos en excel Power BI etc,    deberé cambiar punto por coma

Category_Amount['Amount'] = Category_Amount['Amount'].astype(str).str.replace(".",",")
Category_Amount
Category_Amount.info()

# Como guardar este o cualquier otro DataFrame en un fichero
archivo_csv = "/Users/rogerdefez/Documents/VSCWorkspace/Data-Science-Project/exercices/Category_Amount.csv"
Category_Amount.to_csv(archivo_csv,sep=";",index=True,decimal=",")
print("Guardado con exito", archivo_csv)

archivo_excel = "/Users/rogerdefez/Documents/VSCWorkspace/Data-Science-Project/exercices/Category_Amount.xlsx"
Category_Amount.to_excel(archivo_excel, sheet_name="Agrupacion", header=True,index=False)
print("Guardado con exito", archivo_excel)
```
